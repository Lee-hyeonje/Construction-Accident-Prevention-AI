```python
import pickle
```

```python
with open("범용답변버전V2.pkl", "rb") as f:
    loaded_results = pickle.load(f)
```

```python
loaded_results
```

```python
# loaded_results 데이터를 보기 좋게 출력하는 코드 예시
for idx, item in enumerate(loaded_results):
    print(f"--- 결과 {idx + 1} ---")
    print("질문:")
    print(item["query"])
    print("\n응답:")
    print(item["result"])
    
    print("\n출처 문서:")
    # source_documents가 리스트 형태이고, 각 항목이 Document 객체라고 가정
    for doc in item["source_documents"]:
        # Document 객체는 보통 page_content 속성을 가짐
        print(" -", doc.page_content)
    print("\n" + "="*50 + "\n")
```

```python
# 결과를 저장할 파일 열기 (UTF-8 인코딩)
with open("output.txt", "w", encoding="utf-8") as f:
    for idx, item in enumerate(loaded_results):
        f.write(f"--- 결과 {idx + 1} ---\n")
        f.write("질문:\n")
        f.write(f"{item['query']}\n\n")
        f.write("응답:\n")
        f.write(f"{item['result']}\n\n")
        f.write("출처 문서:\n")
        for doc in item["source_documents"]:
            f.write(f" - {doc.page_content}\n")
        f.write("\n" + "="*50 + "\n\n")
```

```python
only_results = [item["result"] for item in loaded_results if "result" in item]
```

```python
only_results
```

```python
import numpy as np
from sentence_transformers import SentenceTransformer

# Embedding Vector 추출에 활용할 모델(jhgan/ko-sbert-sts) 불러오기
model = SentenceTransformer('jhgan/ko-sbert-sts', use_auth_token=False)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
embedding_model_name = "jhgan/ko-sbert-sts"  # 임베딩 모델 선택
embedding = HuggingFaceEmbeddings(model_name=embedding_model_name)
```

```python
pred_embeddings = embedding.embed_documents(only_results)
print(np.array(pred_embeddings).shape)  # (샘플 개수, 768)
```

```python
import pandas as pd
submission = pd.read_csv('C:/Users/minkyu/Desktop/test/sample_submission.csv', encoding = 'utf-8-sig')
submission.iloc[:,1] = only_results
submission.iloc[:,2:] = pred_embeddings
submission.head()
```
